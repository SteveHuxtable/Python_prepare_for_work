```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# to write a function to visulize the distribution of A and B
def load_df(df_name='data_a.txt'):
    
    df = np.loadtxt(df_name)
    y = df[:, 0]
    X = df[:, 1:]
    new_X = np.zeros((X.shape[0], X.shape[1]+1))
    new_X[:, 0] = 1
    new_X[:, 1:] = X

    return y, new_X
```

```python
# read the two data sets
os.chdir('/Users/stevehu/Documents/python-memo/Python_prepare_for_work/cs229-hw2')
y_A, X_A = load_df('data_a.txt')
y_B, X_B = load_df('data_b.txt')
```

```python
# plot the distributions of the two data sets
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(7, 14))

axes[0].scatter(X_A[:, 1], X_A[:, 2], c=y_A)
axes[1].scatter(X_B[:, 1], X_B[:, 2], c=y_B)

axes[0].set_title('Data set : A')
axes[1].set_title('Data set : B')
```
